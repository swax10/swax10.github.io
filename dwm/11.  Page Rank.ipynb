***Importing Required Libaries***

```python
import numpy as np
import scipy as sc
import pandas as pd
from fractions import Fraction
```

```python
def display_format(my_vector, my_decimal): return np.round((my_vector).astype(np.float), decimals=my_decimal)
my_dp = Fraction(1,3)
Mat = np.matrix([[0, 0, 1],
                 [Fraction(1, 2), 0, 0],
                 [Fraction(1, 2), 1, 0]]).astype(float)
Ex = np.zeros((3,3))
Ex[:] = my_dp
beta = 0.7
Al = beta * Mat + ((1-beta) * Ex)
r = np.matrix([my_dp, my_dp, my_dp])
r = np.transpose(r)
```

```python
previous_r = r  # Initialize previous_r before the loop
for i in range(1, 100):
    r = Al * r
    print(display_format(r, 3))
    if (previous_r == r).all():
        break
    previous_r = r
```

***Final Output***

```python
print("Final:\n", display_format(r, 3))
print("sum", np.sum(r))
```
